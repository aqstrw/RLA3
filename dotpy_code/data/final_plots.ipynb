```python
from utils import read_data, plot_smoothed_scores_wip, smooth
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set hyperparameters here

learnrate = 0.01
gam = 0.99
runs = 8
ep_num = 2000
save_data_cadence = 500
smoothen_over = 11
if save_data_cadence>ep_num:
    print("WARNING : data will not get saved if save cadence is more than ep_num")

# bookkeeping
fname = "reinforce_v1_lr"+"{:.4f}".format(learnrate)[-4:]+"_g"+"{:.4f}".format(gam)[-4:]+"_runs{:1d}_eps".format(runs)+str(ep_num)
    
# read data
score_stack_lr01 = read_data(fname)
    
# plot results
plot_smoothed_scores_wip(score_stack_lr01, sav_window = smoothen_over, avg_plot = 1, save = 0, fname = fname)
```

```python
# set hyperparameters here

learnrate = 0.0001
gam = 0.99
runs = 4
ep_num = 2000
save_data_cadence = 500
smoothen_over = 11
if save_data_cadence>ep_num:
    print("WARNING : data will not get saved if save cadence is more than ep_num")

# bookkeeping
fname = "reinforce_v1_lr"+"{:.4f}".format(learnrate)[-4:]+"_g"+"{:.4f}".format(gam)[-4:]+"_runs{:1d}_eps".format(runs)+str(ep_num)
    
# read data
score_stack_lr0001 = read_data(fname)
    
# plot results
plot_smoothed_scores_wip(score_stack_lr0001, sav_window = smoothen_over, avg_plot = 1, save = 0, fname = fname)
```

```python
# set hyperparameters here

learnrate = 0.001
gam = 0.99
runs = 4
ep_num = 1000
save_data_cadence = 500
smoothen_over = 11

# bookkeeping
fname1 = "reinforce_v1_lr"+"{:.4f}".format(learnrate)[-4:]+"_g"+"{:.4f}".format(gam)[-4:]+"_runs{:1d}_eps".format(runs)+str(ep_num)

learnrate = 0.001
gam = 0.99
runs = 3
ep_num = 1000
save_data_cadence = 500
smoothen_over = 11

fname2 = "reinforce_v1_lr"+"{:.4f}".format(learnrate)[-4:]+"_g"+"{:.4f}".format(gam)[-4:]+"_runs{:1d}_eps".format(runs)+str(ep_num)
# read data

learnrate = 0.001
gam = 0.99
runs = 3
ep_num = 2000
save_data_cadence = 500
smoothen_over = 11

# fname3 = "reinforce_v1_lr"+"{:.4f}".format(learnrate)[-4:]+"_g"+"{:.4f}".format(gam)[-4:]+"_runs{:1d}_eps".format(runs)+str(ep_num)

# read data

score_stack_lr001 = np.vstack([read_data(fname1),read_data(fname2)])
# print(score_stack.shape)
    
# plot results
plot_smoothed_scores_wip(score_stack_lr001, sav_window = smoothen_over, avg_plot = 1, save = 0, fname = fname)
```

```python
##### together

save = 1
sav_window = 11

fig,axs = plt.subplots(1,1,figsize = (7,5))
stacklist = [score_stack_lr01,score_stack_lr001,score_stack_lr0001]
par_list = [0.01,0.001,0.0001]
for i,stack in enumerate(stacklist):
    avg = np.nanmean(stack, axis=0)
    std = np.std(stack, axis=0)
    axs.fill_between(range(len(avg)), smooth(avg - std, sav_window), smooth(avg + std, sav_window), alpha=0.2)
    axs.plot(smooth(avg,sav_window), label="LR = {}".format(par_list[i])+", average over " + str(len(stack)) + " runs")

# set labels
axs.set_xlabel("Episodes")
axs.set_ylabel("Rewards")

# plot legends
axs.legend()

# save plot
if save:
    plt.tight_layout()
    fig.savefig("initial_lr_test.pdf")

# display fig
fig.show()
```
